```python
import os

os.listdir('../data/dataset/train/')
```

```python
path = '../data/dataset/train/'
ls = os.listdir(path)
train_imgs, val_imgs, total_imgs = [], [], []
mn = float('inf')
mx = 0
for i in ls:
    imgs = os.listdir(path+i)
    if len(imgs) > 2:
        val_imgs.append(path+i+'/'+imgs[0])
        val_imgs.append(path+i+'/'+imgs[1])
        for j in range(2, len(imgs)):
            train_imgs.append(path+i+'/'+imgs[j])
    else:
        for j in range(len(imgs)):
            train_imgs.append(path+i+'/'+imgs[j])
    for j in range(len(imgs)):
        total_imgs.append(path+i+'/'+imgs[j])

    
len(train_imgs), len(val_imgs), len(total_imgs)
```

```python
person_to_img = {}

for img in total_imgs:
    name = img.split('/')[-2]
    if name not in person_to_img:
        person_to_img[name] = []

    person_to_img[name].append(img)

persons = list(person_to_img.keys())
len(persons)
```

```python
mn = 99999999999
cnt = 0
for person in persons:
    if len(person_to_img[person]) < 2:
        cnt += 1
    mn = min(mn, len(person_to_img[person]))
mn, cnt
```
